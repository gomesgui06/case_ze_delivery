```python
import pandas as pd
from preprocessing import PreProcessing

pd.set_option('display.max_columns', None)
```

```python
android = 'reviews_googleplay_1623104726.csv'
ios = 'reviews_itunes_1623103926.csv'
raw = f'../Datalake/Raw'
```

```python
df_android = pd.read_csv(f'{raw}/{android}')
df_ios = pd.read_csv(f'{raw}/{ios}')
```

```python
def clean_column(df):
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace('&', '')
    df.columns = df.columns.str.replace(' ', '_', regex=True)
    df.columns = df.columns.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
    return df

df_android = clean_column(df_android)
df_ios = clean_column(df_ios)
```

```python
android_columns = ['date', 'version', 'author', 'rating', 'review']
ios_columns = ['date', 'version', 'author', 'rating', 'review']
```

```python
df_android.head()
```

```python
df_ios.head()
```

```python
df_android = df_android[android_columns]
df_android['os'] = 'android'
df_android.head()
```

```python
df_ios = df_ios[ios_columns]
df_ios['os'] = 'ios'
df_ios.head()
```

```python
df = df_android.append(df_ios)
```
